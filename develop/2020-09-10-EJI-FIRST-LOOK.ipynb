```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
import seaborn as sns 
sns.set; 
```

```python
Df = pd.read_excel('../data/coalpublic2013.xls', header=3, index_col='MSHA ID')
Df.head()
```

```python
Df['Company Type'].unique()
```

```python
Df['Company Type'].replace(to_replace ='Indepedent Producer Operator', value='Independent Produer Operator', inplace=True)
```

```python
Df['Company Type'].head()
```

```python
Df.rename(columns=lambda x: x.replace(' ','_'), inplace=True)
```

```python
Df.head()
```

```python
plt.scatter(Df.Average_Employees, Df.Labor_Hours)
plt.xlabel('Number Employees')
plt.ylabel('Total Labour Hours')
plt.savefig('../figures/ 2020-09-10' + 'Scatterplt_employee_v_Labour_hours.png')
```
